```python
#########"lahne bech nchouf aandichi colonne cv pas unamed walla haja  yaaani colonne vide tabda cache 
import pandas as pd

# 1. S'assurer que cv_users est bien un DataFrame
if isinstance(cv_users, pd.Series):
    cv_users = cv_users.to_frame()

# 2. S'assurer que les colonnes sont bien du texte
cv_users.columns = cv_users.columns.map(str)
offres.columns = offres.columns.map(str)

# 3. Suppression des colonnes inutiles (colonnes 'Unnamed')
colonnes_a_supprimer_offres = [col for col in offres.columns if 'Unnamed' in col]
offres.drop(columns=colonnes_a_supprimer_offres, inplace=True)

colonnes_a_supprimer_cv = [col for col in cv_users.columns if 'Unnamed' in col]
cv_users.drop(columns=colonnes_a_supprimer_cv, inplace=True)

# 4. Suppression des lignes avec trop de valeurs manquantes
offres = offres.dropna(thresh=int(0.5 * offres.shape[1]))
cv_users = cv_users.dropna(thresh=int(0.5 * cv_users.shape[1]))

# 5. Remplir les petites valeurs manquantes restantes
offres = offres.fillna('')
cv_users = cv_users.fillna('')

# 6. Résumé
print("✅ Nettoyage terminé.")

print(f"\nNouvelles dimensions Offres d'emploi : {offres.shape}")
print(f"Nouvelles dimensions CV Utilisateurs : {cv_users.shape}")

print("\nInfos Offres d'emploi après nettoyage :")
offres.info()

print("\nInfos CV Utilisateurs après nettoyage :")
cv_users.info()

print("\nValeurs manquantes restantes dans Offres d'emploi :")
print(offres.isnull().sum())

print("\nValeurs manquantes restantes dans CV Utilisateurs :")
print(cv_users.isnull().sum())
```

```python
#########"" lahne bech naaml transfomration lel competence fi vecteur 
from sklearn.feature_extraction.text import TfidfVectorizer

# Utiliser TfidfVectorizer pour transformer les compétences texte en vecteurs numériques
vectorizer = TfidfVectorizer()

# Transformation des compétences des utilisateurs (CV)
X_cv = vectorizer.fit_transform(cv_users['competences_cv'])

# Transformation des compétences des offres d'emploi
X_offres = vectorizer.transform(offres['competences'])

print("✅ Transformation TF-IDF terminée.")
print(f"Dimensions des CV vectorisés : {X_cv.shape}")
print(f"Dimensions des Offres vectorisées : {X_offres.shape}")
```

```python
############" lahne bech nchouf similaritè bin cv wel offre 
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# Calculer la similarité entre chaque CV et chaque Offre
similarites = cosine_similarity(X_cv, X_offres)

# Pour chaque CV, trouver l'offre avec la meilleure similarité
meilleures_offres_indices = np.argmax(similarites, axis=1)

# Récupérer les infos de l'offre la plus proche pour chaque CV
offres_recommandees = offres.iloc[meilleures_offres_indices]

# Ajoutons les recommandations aux utilisateurs
cv_users['offre_recommandee_id'] = offres_recommandees['id'].values
cv_users['offre_recommandee_titre'] = offres_recommandees['titre'].values
cv_users['offre_recommandee_competences'] = offres_recommandees['competences'].values

print("✅ Matching CV ↔ Offres terminé avec succès !")
display(cv_users[['nom', 'competences_cv', 'offre_recommandee_titre', 'offre_recommandee_competences']].head())
```

```python
from sklearn.model_selection import train_test_split

# Nouvelle cible : offre recommandée par son titre
y = cv_users['offre_recommandee_titre']

# Nouvelle division
X_train, X_test, y_train, y_test = train_test_split(X_cv, y, test_size=0.2, random_state=42)

print("✅ Nouveau jeu de données prêt !")
print(f"X_train shape : {X_train.shape}")
print(f"X_test shape  : {X_test.shape}")
print(f"y_train length : {len(y_train)}")
print(f"y_test length  : {len(y_test)}")
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report

# Nouveau modèle KNN
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)

# Prédictions
y_pred = knn.predict(X_test)

# Évaluation
accuracy = accuracy_score(y_test, y_pred)
print(f"✅ Nouvelle Accuracy du KNN : {accuracy:.4f}")

print("\nClassification Report KNN :")
print(classification_report(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

# Création du modèle Random Forest
rf = RandomForestClassifier(n_estimators=100, random_state=42)

# Entraînement du modèle
rf.fit(X_train, y_train)

# Prédictions
y_pred_rf = rf.predict(X_test)

# Évaluation
accuracy_rf = accuracy_score(y_test, y_pred_rf)
print(f"✅ Accuracy du modèle Random Forest : {accuracy_rf:.4f}")

# Rapport détaillé
print("\nClassification Report Random Forest :")
print(classification_report(y_test, y_pred_rf))
```

```python
from sklearn.preprocessing import LabelEncoder
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, classification_report

# 1. Encoder les labels (transformer les textes en nombres)
label_encoder = LabelEncoder()
y_train_encoded = label_encoder.fit_transform(y_train)
y_test_encoded = label_encoder.transform(y_test)

# 2. Création du modèle XGBoost
xgb = XGBClassifier(use_label_encoder=False, eval_metric='mlogloss', random_state=42)

# 3. Entraînement du modèle
xgb.fit(X_train, y_train_encoded)

# 4. Prédictions
y_pred_xgb = xgb.predict(X_test)

# 5. Décoder les prédictions pour retrouver les noms
y_pred_labels = label_encoder.inverse_transform(y_pred_xgb)

# 6. Évaluation
accuracy_xgb = accuracy_score(y_test, y_pred_labels)
print(f"✅ Accuracy du modèle XGBoost : {accuracy_xgb:.4f}")

# Rapport détaillé
print("\nClassification Report XGBoost :")
print(classification_report(y_test, y_pred_labels))
```
