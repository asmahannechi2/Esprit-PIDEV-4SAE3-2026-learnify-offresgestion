# 🎓 Learnify English : Moteur Recrutement IA (V2)
Ce notebook combine **Segmentation** et **Recommandation** avec une spécialisation totale pour l'enseignement de l'Anglais.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.ensemble import RandomForestClassifier

print("✨ Environnement Learnify IA optimisé.")
```

## 1. Chargement et Enrichissement des Données
Ici, nous ajoutons des mots-clés sémantiques pour garantir qu'un profil d'enseignant ne tombe jamais à 0% de match.

```python
# Simulation de données spécialisées
data_offres = {
    'titre': ['English Teacher', 'ESL Instructor', 'IELTS Specialist', 'Business English Coach', 'Phonics Teacher', 'Young Learners Tutor'],
    'competences': ['Grammar, TEFL, Speaking', 'Adult Education, CELTA, Communication', 'Academic English, IELTS, TOEFL', 'ESP, Corporate, Business writing', 'Kids, Montessori, Phonics', 'Creative writing, Games, Elementary']
}
offres = pd.DataFrame(data_offres)

data_cv = {
    'nom': ['Amine', 'Sarah', 'Karim', 'Linda'],
    'cv_text': [
        'I have a CELTA and experience in teaching kids',
        'IELTS expert with 10 years in academic writing',
        'Business background, now teaching English to managers',
        'Beginner teacher with TEFL certification'
    ]
}
cv_df = pd.DataFrame(data_cv)

# Enrichissement sémantique (La clé pour éviter les scores 0)
def semantic_boost(text):
    return f"{text} english teacher education language instructor"

offres['vector_text'] = offres.apply(lambda b: f"{b['titre']} {b['competences']}", axis=1).apply(semantic_boost)
cv_df['vector_text'] = cv_df['cv_text'].apply(semantic_boost)

print("📊 Données enrichies.")
```

## 2. Recommendation Engine (Matching)
Calcul de la similarité cosinus avec un garde-fou pour les résultats.

```python
vectorizer = TfidfVectorizer(stop_words='english')
X_offres = vectorizer.fit_transform(offres['vector_text'])
X_cv = vectorizer.transform(cv_df['vector_text'])

similarities = cosine_similarity(X_cv, X_offres)

print("✅ Matrice de recommandation calculée.")
results = pd.DataFrame(similarities, index=cv_df['nom'], columns=offres['titre'])
sns.heatmap(results, annot=True, cmap='YlGnBu')
plt.title('Matching Scores (%)')
plt.show()
```

## 3. Segmentation Automatique
Classification des candidats par domaine d'expertise.

```python
# On définit les catégories (Segments)
segments = ['General ESL', 'Academic/IELTS', 'Business English', 'Young Learners']
y = [0, 1, 2, 0] # Catégories pour nos 4 exemples de test

clf = RandomForestClassifier()
clf.fit(X_cv, y)

def classify_cv(new_text):
    vec = vectorizer.transform([semantic_boost(new_text)])
    pred = clf.predict(vec)[0]
    return segments[pred]

test_cv = "I want to teach managers and business professionals"
print(f"🔍 Segmentation pour le nouveau CV : {classify_cv(test_cv)}")
```
